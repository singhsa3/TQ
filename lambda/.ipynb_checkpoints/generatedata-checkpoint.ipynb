```python
#export
import pickle5 as pickle
import math
import sys
import time
import copy
import numpy as np
import pandas as pd
import librosa, librosa.display
import matplotlib.pyplot as plt

import os
import glob

import cv2
import pickle5 as pickle
from PIL import Image as im
from joblib import Parallel, delayed
from os.path import exists
import pandas as pd

pathG='../data' 
DATASET_PATH0 = pathG+"/w2v2/w2v2L0"
DATASET_PATH4 = pathG+"/w2v2/w2v2L4"
DATASET_PATH8 = pathG+"/w2v2/w2v2L8"
DATASET_PATHS = [DATASET_PATH0,DATASET_PATH4,DATASET_PATH8]
filenames= glob.glob(DATASET_PATHS[0]+"/*.pickle" )
filenames = [os.path.basename(x) for x in filenames]
def getZ(x): 
    c=np.array(x[0]).T    
    #print(c.shape)
    b = np.max(c)
    a = np.min(c)
    c =(c-a)/(b-a)
    c= pd.DataFrame(c)
    #c= c.astype('float64')
    #print(type(c))
    c=c.describe(percentiles=[.05,.1,.2,.3,.4,.5,.6,.7,.8,.9,.95])
    #print(c.shape)
    c=c.drop(['count'])
    c=np.array(c.T)
    c=c.reshape(1,c.shape[0],c.shape[1])
    return  c
def db1(DATASET_PATH,recrd):
    file = DATASET_PATH +"/"+ recrd
    flo = recrd.split('.')[0]
    fl =DATASET_PATH+'/'+flo+'.pickle'
    fln=DATASET_PATH+'/db1/'+flo+'.pickle'
    if exists(fln)==False:
        try:
            with open(fl,"rb") as f:
                x=pickle.load(f)
            x=getZ(x)
            with open(fln, 'wb') as handle:
                pickle.dump(x,handle, protocol=pickle.HIGHEST_PROTOCOL)
        except Exception as e:
            print(str(e))
            
results = Parallel(n_jobs=-1)(delayed(db1)(DATASET_PATH,recrd) for recrd in filenames for DATASET_PATH in DATASET_PATHS)             
    
```
